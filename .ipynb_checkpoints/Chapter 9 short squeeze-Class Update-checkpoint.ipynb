```python
from IPython.display import display, Math, Latex

import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#read in share price history for Game Stop from January, 2021 to February 02, 2021
share = yf.Ticker('GME')
start_date = '2021-01-04'
end_date = '2021-02-15'
hist = share.history(start=start_date, end=end_date)
hist
```

```python
prices = pd.DataFrame(hist['Close'])
volume = pd.DataFrame(hist['Volume'])
#prices.head()
#volume.head()
```

```python
#Graph the stock price of Gamestop over the full range of dates
plt.plot(prices, color='b')
plt.title('Gamestop closing prices')
plt.xticks(rotation=-90)
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
#create a bar chart for daily volume

plt.bar(volume.index.get_level_values('Date'), volume.iloc[:,0])
plt.title('Gamestop daily volume')
plt.xlabel('Date')
plt.xticks(rotation=-90)
plt.ylabel('Volume')
plt.show()
```
